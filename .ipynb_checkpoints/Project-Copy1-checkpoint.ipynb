```python
## https://stackoverflow.com/questions/50996820/how-to-import-a-sqlite3-database-into-python-jupyter-notebook

albums = pd.read_sql_query("SELECT * FROM albums", cnx)
albums.head(50)
```

```python
## https://stackoverflow.com/questions/50996820/how-to-import-a-sqlite3-database-into-python-jupyter-notebook

feat = pd.read_sql_query("SELECT * FROM acoustic_features", cnx)
feat.head(50)
```

```python
feat.describe()
```

```python
feat.date.astype(str)
albums.date.astype(str);
```

```python
del feat['id']
```

```python
fulldata = pd.merge(albums, feat, how = "outer")
fulldata.head(100)
```

```python
fulldata.describe()
```

```python
albums[albums['length'] > 200]
```

```python
# all albums that were number one each day
num1 = fulldata[fulldata['rank'] == '1'].copy()
num1.head(20)
# df_names = {k: v for (k, v) in df.groupby('customer name')}
```

```python
# number of days each album was number one 
freq1 = num1.groupby(['artist','album'])[['id']].count()
freq1
# histogram of number of days albums were number one?
```

```python
#frequency grester than 1
freqgt1 = freq1[freq1['id']>1]
freqgt1
```

```python
#frequency grester than 1
freqgt2 = freq1[freq1['id']>2]
freqgt2
```

```python
# https://www.geeksforgeeks.org/python-pandas-dataframe-drop_duplicates/
num1unique = num1.copy()
num1unique.drop_duplicates(subset ="album", 
                     keep = 'first', inplace = True)
num1unique.head(20)
```
